```python
df = df.with_columns(pl.col("pchembl_value_right").alias("pchembl_value"))
```

```python
import polars as pl

# Ostrze┼╝enie: Wykonaj to na surowym df wczytanym do pami─Щci z Twojego oryginalnego parquetu.
df_cleaned = (
    df
    # 1. Usuni─Щcie oznacze┼Д niewa┼╝no┼Ыci wprowadzonych przez rzeczoznawc├│w ChEMBL (np. "Outside typical range")
    .filter(pl.col("data_validity_comment").is_null())
    
    
    # 3. Zast─Еpienie pustej kolumny pchembl_value wersj─Е zawieraj─Еc─Е w┼Вa┼Ыciwe warto┼Ыci
    .drop(["pchembl_value"])
    .rename({"pchembl_value_right": "pchembl_value"})
    
    # 4. Rzutowanie typ├│w tekstowych na liczne, ignorujemy b┼В─Щdy (strict=False spowoduje, ┼╝e nierozpoznane warto┼Ыci zamieni─Е si─Щ w Nulle)
    .with_columns([
        pl.col("pchembl_value").cast(pl.Float32, strict=False),
        pl.col("psa").cast(pl.Float32, strict=False),
        pl.col("mw").cast(pl.Float32, strict=False),
        pl.col("hba").cast(pl.Int32, strict=False),
        pl.col("hbd").cast(pl.Int32, strict=False)
    ])
    
    # 5. Ca┼Вkowite odrzucenie wierszy w kt├│rych brakuje kluczowych warto┼Ыci po transformacjach
    # fingerprint usuni─Щte z tej konkretnej regu┼Вy, gdy┼╝ czasem fingerprint dodajesz p├│┼║niej. 
    # Zostawmy upewnienie si─Щ co do pChEMBL, smiles i masy
    .drop_nulls(subset=["pchembl_value", "canonical_smiles", "mw"])
)

print(f"Liczno┼Ы─З przed czyszczeniem: {df.shape[0]}")
print(f"Liczno┼Ы─З po kompleksowym czyszczeniu: {df_cleaned.shape[0]}")
print(f"Zredukowano zbi├│r o {df.shape[0] - df_cleaned.shape[0]} w─Еtpliwych lub pustych rekord├│w.")
```

```python
# Szybkie sprawdzenie poprawno┼Ыci wynik├│w:
display(df_cleaned.select(pl.col("pchembl_value")).describe())

# Pozbycie si─Щ resztek _right w nazwach kolumn tak aby utrzyma─З zbi├│r czysty (opcjonalne)
columns_to_drop = [col for col in df_cleaned.columns if col.endswith("_right")]
df_cleaned = df_cleaned.drop(columns_to_drop)
print(f"Brakuj─Еce kolumny po sprz─Еtaniu sufiks├│w '_right': {columns_to_drop}")
```

```python
import polars as pl

# Zak┼Вadamy, ┼╝e masz ju┼╝ czysty "df_cleaned" gdzie wyfiltrowano z┼Вe nulle

df_deduplicated = (
    df_cleaned
    # 1. Obieramy sobie unikatowy ┼Вa┼Дcuch molekularny za fundament (ka┼╝dy unikatowy SMILES b─Щdzie reprezentowa┼В dok┼Вadnie 1 docelowy wiersz)
    .group_by("canonical_smiles")
    # 2. Narzucamy zasady, co zrobi─З z dziesi─Еtkami kolumn duplikat├│w:
    .agg([
        # Dla warto┼Ыci pomiarowych (targetu iIC50/pChEMBL) bierzemy MEDIAN─Ш (jest wysoce odporna na odskakuj─Еce, ob┼В─Щdne b┼В─Щdy i daje solidny werdykt ze wszystkich laboratori├│w)
        pl.col("pchembl_value").median().alias("pchembl_value"),
        
        # W┼Вa┼Ыciwo┼Ыci chemiczne (masa g─Щsto┼Ы─З itp.) zawsze b─Щd─Е takie same dla tej samej cz─Еsteczki, wi─Щc pobieramy po prostu jej pierwsz─Е warto┼Ы─З ze spi─Щtrzonego stosu zduplikowanych zg┼Вosze┼Д 
        pl.col("mw").first(),
        pl.col("psa").first(),
        pl.col("hbd").first(),
        pl.col("hba").first(),
        pl.col("fingerprint").first(),      # Je┼╝eli jest ju┼╝ wygenerowany, to pierwszy z g├│ry b─Щdzie OK!
        
        # Opcjonalnie: Zliczenie ile duplikat├│w zebra┼Вo si─Щ na ten jeden wynik (na wypadek ch─Щci analizy rozbie┼╝no┼Ыci, kto by chcia┼В)
        pl.col("pchembl_value").count().alias("measurements_merged_count")
    ])
)

print(f"Stworzono doskona┼Вy zbi├│r gotowy dla Modelu AI zredukowany do {df_deduplicated.shape[0]} w pe┼Вni unikalnych atomowo rz─Щd├│w!")
```

```python
target_smiles = "COc1cc2ncnc(Nc3ccc(F)c(Cl)c3)c2cc1OCCCN1CCOCC1"

# Z┼Вap z surowego df tylko te rz─Щdy (badania), gdzie kszta┼Вt to nasza cz─Еsteczka
gefitinib_records = df.filter(pl.col("canonical_smiles") == target_smiles)

# Poka┼╝ jak du┼╝o dokument├│w o tym pisze i wy┼Ыwietl ca┼В─Е tabel─Щ:
print(f"Ilo┼Ы─З r├│┼╝nych bada┼Д wok├│┼В tej pojedynczej cz─Еsteczki: {gefitinib_records.shape[0]}")

# Pokazanie wszystkich kolumn, ┼╝eby zobaczy─З dok┼Вadne r├│┼╝nice
pl.Config.set_tbl_cols(100) # Po to, aby nie poukrywa┼Вo Ci ┼Ыrodk├│w tabel skracaj─Еc kolumny

display(gefitinib_records)
```
