```python
import polars as pl


df_deduplicated = (
    df
    # 1. Obieramy sobie unikatowy łańcuch molekularny za fundament (każdy unikatowy SMILES będzie reprezentował dokładnie 1 docelowy wiersz)
    .group_by("canonical_smiles")
    # 2. Narzucamy zasady, co zrobić z dziesiątkami kolumn duplikatów:
    .agg([
        # Dla wartości pomiarowych (targetu iIC50/pChEMBL) bierzemy MEDIANĘ (jest wysoce odporna na odskakujące, obłędne błędy i daje solidny werdykt ze wszystkich laboratoriów)
        pl.col("pchembl_value").median().alias("pchembl_value"),
        
        # Właściwości chemiczne (masa gęstość itp.) zawsze będą takie same dla tej samej cząsteczki, więc pobieramy po prostu jej pierwszą wartość ze spiętrzonego stosu zduplikowanych zgłoszeń 
        pl.col("mw").first(),
        pl.col("psa").first(),
        pl.col("hbd").first(),
        pl.col("hba").first(),
        pl.col("ecfp_smiles").first(),      # Jeżeli jest już wygenerowany, to pierwszy z góry będzie OK!
        
        # Opcjonalnie: Zliczenie ile duplikatów zebrało się na ten jeden wynik (na wypadek chęci analizy rozbieżności, kto by chciał)
        pl.col("pchembl_value").count().alias("measurements_merged_count")
    ])
)

print(f"Stworzono doskonały zbiór gotowy dla Modelu AI zredukowany do {df_deduplicated.shape[0]} w pełni unikalnych atomowo rzędów!")
```
